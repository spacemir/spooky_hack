This is a first test using scikit-learn

```python
import  pprint
import  subprocess 
import sys 
sys.path.append('../')

import numpy as np
import scipy as sp
import statsmodels.api as sm
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns

import pandas
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

# Train

```python
data_path = './train.csv'
data = pandas.read_csv(data_path)
data
```

## Vectorizer for bag of words

```python
Y = data['author']
text = data['text']
all_text = []
for x in text:
    all_text.append(x)
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(all_text)
```

#### Normalize

```python
tfidf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
```

```python
X_train_tfidf = tfidf_transformer.transform(X_train_counts)
```

#### train

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, Y)
```

```python
clf.score(X_train_tfidf, Y)
```

```python
clf.classes_
```

# Predict

```python
prediction_path = './test.csv'
prediction_data = pandas.read_csv(prediction_path)
prediction_data
```

```python
text = prediction_data['text']
prediction_text = []
for x in text:
    prediction_text.append(x)
```

#### Pre-process

```python
X_test_counts = count_vect.transform(prediction_text)
X_text_counts_tf_idf = tfidf_transformer.transform(X_test_counts)
```

#### Predict

```python
clf.predict_proba(X_text_counts_tf_idf)
```

## Save submission

```python
submission_path = './sample_submission.csv'
submission_data = pandas.read_csv(submission_path)
submission_data
```

```python
df = pandas.DataFrame(columns=["id", "EAP", "HPL", "MWS"])
df["id"] = submission_data['id']
df["EAP"] = clf.predict_proba(X_text_counts_tf_idf)[:, 0]
df["HPL"] = clf.predict_proba(X_text_counts_tf_idf)[:, 1]
df["MWS"] = clf.predict_proba(X_text_counts_tf_idf)[:, 2]
```

```python
df.to_csv('./prediction.csv', index=False)
```
